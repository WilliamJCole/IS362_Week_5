##### Calculating the northernmost Airport.
Lets start off by importing the file and display the information by locations.

```python
airports = pd.read_csv("https://raw.githubusercontent.com/hadley/nycflights13/master/data-raw/airports.csv")
airports.sort_values(by=["lat"], ascending=False)
```

The informations shows that the most northern airport is *Diliant Hopkins*. This however is not correct, the lat and lon was inverted and the **Wiley Post Will Rogers Airport** is infact the most notherern airport.

###### Calculating the easternmost airport.

Let's now check out the eastern most Airport:

```python
df = pd.read_csv("https://raw.githubusercontent.com/hadley/nycflights13/master/data-raw/airports.csv")
east_airports = df[(df['lat'] > 44) & (df['lat'] <= 50) & (df['lon'] >= -75) & (df['lon'] <= -65)]
east_airports.sort_values('lon', ascending=False)
```

The *Eastport Municipal Airport* is the eastern most airport. I have confirmed this on Google maps.
